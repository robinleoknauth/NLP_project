```python
def get_recommendations(first_article,recommendation_number ,topics,ted_data, model, vectorizer, training_vectors):
    
    new_vec = model.transform(
        vectorizer.transform([first_article]))
    
    nn = NearestNeighbors(n_neighbors=recommendation_number, metric='cosine', algorithm='brute')
    nn.fit(training_vectors)
    
    results = nn.kneighbors(new_vec)
    
    recommend_list = results[1][0]
    scores = results[0]
                       
    ss = np.array(scores).flat
    for i, resp in enumerate(recommend_list):
        print('\n--- TED ID ---\n', + resp)
        print('\n--- TED Title ---')
        print(ted_data.iloc[resp]['title'])
#         print('\n')
        print('--- Distance from original ---\n', + ss[i]) 
        print('\n')
        print('--- Topics ---')
        print(topics.iloc[resp,0])
        print(ted_data.iloc[resp,1])
        print('\n')
        print('--- offical ted tags ---')
        print(ted_data.iloc[resp]['tags'])
        print('___________________________________________\n')
#         print(resp)
    return recommend_list, ss
```
